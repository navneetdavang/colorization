### Importing the Library

```python
import numpy as np
import matplotlib.pyplot as plt 
from skimage.io import imread, imshow
from skimage.color import rgb2lab, lab2rgb, rgb2gray, gray2rgb
from skimage.transform import resize
import tensorflow as tf
import tensorflow_io as tfio
```

#### Image File Path

```python
imgPath = '/content/test4.jpg'
```

### skimage Image reading 

```python
skimage = imread(imgPath)
```

```python
imshow(skimage)
```

#### tf Image reading

```python
_tfImg = tf.io.read_file(imgPath)
```

```python
_tfImg
```

```python
# decoding the image
tfImg = tf.image.decode_jpeg(_tfImg)
```

```python
tfImg
```

Image Read Comparision

#### LAB Conversion

##### Using the skimage

```python
skilab = rgb2lab(skiResize)
```

```python
skilab
```

```python
imshow(skilab)
```

```python
print('skilab status :')
print(f'0 min : {np.min(skilab[:,:,0])}')
print(f'0 max : {np.max(skilab[:,:,0])}')
print(f'1 min : {np.min(skilab[:,:,1])}')
print(f'1 max : {np.max(skilab[:,:,1])}')
print(f'2 min : {np.min(skilab[:,:,2])}')
print(f'2 max : {np.max(skilab[:,:,2])}')
```
